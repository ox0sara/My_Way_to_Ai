$\hat{y} = x_1 \cdot w_1 + x_2 \cdot w_2$

```python
X_train = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1] ]

Y_train = [0, 1, 1, 1]
```

```python
# activation function
def f(z):
    return 1 if z > 0 else 0


# training function
def fit(X_train, Y_train, num_iterations):
    w1, w2 = 0, 0   # initialize weights

    for iteration in range(num_iterations):
        correct = 0

        for x, y in zip(X_train, Y_train):
            x1, x2 = x

            # prediction
            y_hat = f(x1 * w1 + x2 * w2)

            # count correct predictions
            if y == y_hat:
                correct += 1

            # compute error
            d = y - y_hat

            # update weights
            w1 = w1 + d * x1
            w2 = w2 + d * x2
            print(f" Expected: {y}, but was :{y_hat}, new w1 :{w1}, new w2: {w2}")
    print(f"correct: {correct} /{len(Y_train)}")
```

```python
fit(X_train, Y_train, 2)
```
